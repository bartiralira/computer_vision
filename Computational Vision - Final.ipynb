## 4.5 Indicadores de desempenho

O Keras não possui os indicadores de desempenho como precisão, sensibilidade e pontuação f1 por padrão, portanto precisamos implementar externamente.

````python
def recall_score(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def precision_score(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision

def f1_score(y_true, y_pred):
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))
````

### 4.5.1 Arquitetura ResNet50

**Pergunta**: Explique como é constituída a arquitetura do ResNet50? *Utilize, se necessário, gráficos, projetos que utilizam essa arquitetura. Detalhe também sua topologia em camadas e mostre quais as situações essa arquitetura pode ter mais êxito e quais cenários não tem.*

**Resposta**: Para entender a arquitetura ResNet50 é primeiro necessário entender o que é a arquitetura ResNet.

**ResNet**

ResNet foi apresentada na ILSVRC 2015, a qual ela venceu com uma performance no top-5 error rate de 3,57%. A Rede Neural Residual (ResNet) apresentada por Kaiming He et al introduziram uma nova arquitetura com duas características principas:

* “identity shortcut connections”: uma estratégia de “atalhos” ou “conexões de salto” que pulam pares de grupos de camadas convolucionais. São também chamadas gated units ou gated recurrent units, apesar de não apresentarem uma recorrência no sentido tradicional dos modelos de redes neurais recorrentes;

* foco pesado em normalização de lotes (batch normalization).

Graças a este enfoque eles foram capazes de treinar uma rede com até 152 camada, mas que possuía uma complexidade final menor do que VGG.

**Conexões de atalho de identidade (Identity shortcut connections)**

Identity shortcut connections.png

De forma simplificada, a ideia dos atalhos em ResNet é evitar que a rede, muito profunda, morra por evanescimento de gradientes através do empilhamento de mapeamentos de identidades que, do ponto de vista matemático, segundo os autores, estão simplesmente empilhando camadas que nao fazem nada. Como isso, como mostra a figura acima, ResNet utiliza, num determinado ponto, um sinal que é a soma do sinal produzido pelas duas camadas convolucionais anteriores somado ao sinal transmitido diretamente do ponto anterior a estas camadas, juntando um sinal processsado com um sinal de uma etapa anterior no processamento. 

**ResNet50**

Entendendo mais sobre o ResNet agora podemos falar um pouco mais sobre o ResNet50, esta arquitetura como o nome indica utiliza 50 camadas  que recebem como entrada uma imagem e é composta por uma convolução com filtro (7x7) seguida de uma série de convoluções com filtros (1x1) e (3x3) terminando com uma camada totalmente conectada determinando o número da predição foliar.

**Topologia odo ResNet50**

resnet50.png

**Utilização**
* Essa arquitetura pode ser usada em tarefas de visão computacional, como classificação de imagem, localização de objetos, detecção de objetos.

* e essa estrutura também pode ser aplicada a tarefas não-visão computacional para lhes dar o benefício da profundidade e reduzir a despesa computacional também.

A técnica de transfer learning consiste de utilizar o mesmo modelo e treiná-lo para outas imagens. Por tal motivo, excluímos a última camada para modelar com as classes que definimos, ou seja, **controle**, **bacteriana** e **viral**.

Informe a quantidade de classes a serem classificadas.

````python
## IMPLEMENTE
qtde_classes = 3
````

````python
conv_base = ResNet50(include_top=False)

for layer in conv_base.layers:
    layer.trainable = False
````

````python
x = conv_base.output
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(128, activation='relu')(x) 

predictions = layers.Dense(qtde_classes, activation='softmax')(x)
model = Model(conv_base.input, predictions)

model.summary()
````

````python
optimizer = optimizers.Adam()
model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=[precision_score, recall_score, f1_score])
````

O número de épocas define quantas vezes o modelo irá treinar e validar o erro, assim ajustando os pesos para melhor convergência.
Escolha o número adequado de épocas para alcançarmos pelo menos **70% de precisão de validação**.

````python
## IMPLEMENTE
qtde_epocas = 10
````

````python
history = model.fit(train_generator, epochs=qtde_epocas, validation_steps=5, steps_per_epoch=5, 
                    validation_data=val_generator)
````

Um modelo que converge bem possui o gráfico de perda (*loss*) descendente e os gráfico de precisão (*precision*), sensibilidade (*recall*) e pontuação f1 (*f1 score*) em acendente.

````python
# Exibindo dados de Precisão
plt.plot(history.history['precision_score'])
plt.plot(history.history['val_precision_score'])
plt.title('model precision')
plt.ylabel('precision')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# Exibindo dados de Sensibilidade
plt.plot(history.history['recall_score'])
plt.plot(history.history['val_recall_score'])
plt.title('model recall')
plt.ylabel('recall')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# Exibindo dados de F1 Score
plt.plot(history.history['f1_score'])
plt.plot(history.history['val_f1_score'])
plt.title('model f1_score')
plt.ylabel('f1_score')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# Exibindo dados de Perda
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
````

A arquitetura de rede VGG utiliza camadas convolucionais 3 x 3 empilhadas umas

Simonyan e Zisserman (2014) foi investigaram o efeito da profundidade da rede

VGG16.PNG

A tabela acima, mostra como é a arquitetura da rede VGG16. Nela é possível ver 13 camadas que fazem convolução retirando as features das imagens e mais 3 camadas totalmente conectadas para fazer as classificações.

Simonyan e Zisserman (2014) acharam o treinamento VGG16 desafiador especificamente em relação à convergência nas redes mais profundas. Então, para facilitar o treinamento, eles primeiro treinaram versões menores de VGG com 11 camadas que serviram como inicializações para as redes maiores e mais profundas como a VGG16.

**Casos de uso e implementação**

Infelizmente, há duas grandes desvantagens com o VGGNet:

1. É dolorosamente lento para treinar.
2. Os próprios pesos da arquitetura de rede são bastante grandes (no que diz respeito ao disco/largura de banda).

Devido à sua profundidade e número de nódulos totalmente conectados, o VGG16 tem mais de 533MB. Isso torna a implantação do VGG uma tarefa cansativa. O VGG16 é usado em muitos problemas de classificação de imagem de aprendizagem profunda; no entanto, arquiteturas de rede menores são muitas vezes mais desejáveis (como SqueezeNet, GoogLeNet, etc.). Mas é um grande bloco de construção para fins de aprendizagem, pois é fácil de implementar.

**Arquitetura**

Uma imagem RGB fixa de (224 * 224) foi dada como entrada para esta rede, o que significa que a matriz era de forma (224.224,3). <br>
O único pré-processamento que foi feito é que eles subtraíram o valor médio de RGB de cada pixel, computado ao longo de todo o conjunto de treinamento.
Núcleos usados de tamanho (3 * 3) com um tamanho de passo de 1 pixel, isso permitiu que eles cobrissem toda a noção da imagem. <br>
Preenchimento espacial foi usado para preservar a resolução espacial da imagem.
o pooling máximo foi realizado em uma janela de 2 * 2 pixels com sride 2. <br>
Isso foi seguido pela unidade linear retificada (ReLu) para introduzir a não linearidade para fazer o modelo classificar melhor e melhorar o tempo computacional, pois os modelos anteriores usavam funções de tanh ou sigmoid, isso se mostrou muito melhor do que aqueles.<br>
